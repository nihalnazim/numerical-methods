# Solving a system of equations using Python

Using the package `sympy`.

This is solving the system of equations from a problem given by Prof. Liang to 114 students. (Problem Set 5)

The system of equations is: 

1. $V-V_1-V_3=0$
2. $V-V_2-V_4=0$
3. $V-V_1-V_5-V_4=0$
4. $V_3-V_1+V_5=0$
5. $2V_4-V_5-V_2=0$

```python
import sympy as sp
#must define variables as symbols
V1, V2, V3, V4, V5 = sp.symbols('V1 V2 V3 V4 V5')
C = sp.symbols('C', nonnegative=True)  # C >= 0
V = sp.symbols('V', positive=True)     # V > 0
```

```python
#define system of equations
#note that we cannot use equal signs in our equations, becuase that = is hard coded into Python for variable assignment
#Instead, sympy automatically assumes an equation is equal to zero unless we explicitly tell it otherwise. So, for example, the equation Q1 + Q2 + Q5 = 0 is written as Q1 + Q2 + Q5, and the equation Q2 + Q3 = Q4 + Q2 is written as sp.Eq(Q2 + Q3, Q4 + Q2)
# below, a set of equations is defined as a list of sympy expressions. 
#The first three equations are written in the form of an expression equal to zero, 
# and the last two equations are written using the sympy Eq function to explicitly state the equality.
equations = [V-V1-V3,
    V-V2-V4,
    V-V1-V5-V4,
    sp.Eq(V3-V1+V5,0),
    sp.Eq(2*V4-V5-V2,0)]
```

```python
sp.solve(equations, (V1, V2, V3, V4, V5), set=True)
```
